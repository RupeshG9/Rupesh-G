# 3.	Write a program to accept 2 lists from the user of 5 nos each . Assume each list will have unique nos. Now find out how many items in these lists are common .

```python
a=[]
b=[]
a = [int(item) for item in input("Enter the list items : ").split()]
b = [int (item) for item in input("Enter the list items : ").split()]
print(a)
print(b)
 
count=len(set(a).intersection(b))
print("The number of common elements in the two lists are: ")
print(count)
```

# 4.	Rewrite the previous code so that your code alos displays the items which are common in both the lists

```python
a=[]
b=[]
a = [int(item) for item in input("Enter the list items : ").split()]
b = [int(item) for item in input("Enter the list items : ").split()]
print(a)
print(b)
 
def common(a, b):
    c = [i for i in a if i in b]
    return c
d=common(a,b)
print("The common elements in the two lists are: ")
print(d)
```
